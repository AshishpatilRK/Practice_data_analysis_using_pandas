### Q1. Find the null values in the dataset and fill null valur that column.

```python
cars.isnull().sum()    # give null values in every variable
```

```python
cal = ['Make','Model','Type','Origin','DriveTrain','MSRP','Invoice']  # list of categorical varibles
```

```python
num = ['EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']# list of numerical variables
```

```python
for i in cal:
    cars[i].fillna(cars[i].mode()[0], inplace=True)    #  use for loop for fill categorical varibles using mode
```

```python
cars[num] = cars[num].fillna(cars[num].mean())   # fill numerical variables using mean
```

```python
cars.isnull().sum()   #  give null values in every variable
```
